```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
marketing = pd.read_csv('C:/Users/Asus/.cache/kagglehub/datasets/janiobachmann/bank-marketing-dataset/versions/1/bank.csv')
marketing
```

```python
#deposit,Ã¼rÃ¼nÃ¼ satÄ±n alÄ±p almadÄ±ÄŸÄ±nÄ± ifade ediyor.Bizim bulmak istediÄŸimiz
marketing['deposit']=marketing['deposit'].map({'yes': 1, 'no': 0})

# BaÄŸÄ±msÄ±z ve baÄŸÄ±mlÄ± deÄŸiÅŸkenleri ayÄ±rma
X = marketing.drop('deposit', axis=1)  # TÃ¼m sÃ¼tunlardan 'deposit' sÃ¼tununu Ã§Ä±karÄ±r.
y = marketing['deposit']              # 'deposit' sÃ¼tununu hedef olarak seÃ§ilir.Ã§

X = X.drop('duration', axis=1)   #duration'u da Ã§Ä±karÄ±yorum.Ã¶zel inceleme gerektiriyor.

# Kategorik sÃ¼tunlarÄ± one-hot encoding ile dÃ¶nÃ¼ÅŸtÃ¼rme
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# Lojistik regresyon modelini oluÅŸturup ve eÄŸitme
log_reg = LogisticRegression(solver='liblinear',max_iter=1000,random_state=42)
```

```python
log_reg.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
log_reg.predict(X_test)
```

```python
# Test seti Ã¼zerinde tahminler.
y_pred_log_reg = log_reg.predict(X_test)

# Model performansÄ±nÄ± deÄŸerlendirme.BÃ¶ylece modelin ne kadar doÄŸru tahmin yaptÄ±ÄŸÄ±nÄ± anlayabiliriz.
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
report_log_reg = classification_report(y_test, y_pred_log_reg)

print("Accuracy(DoÄŸruluk):", accuracy_log_reg)
print("Classification Report:\n", report_log_reg)

#support:sÄ±nÄ±fÄ±n gerÃ§ek Ã¶rnek sayÄ±sÄ±.
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Tahmin edilen sÄ±nÄ±flar ve gerÃ§ekteki sÄ±nÄ±flar arasÄ±ndaki karÅŸÄ±laÅŸtÄ±rmayÄ± yapan "karÄ±ÅŸÄ±klÄ±k matrisi"ni oluÅŸturalÄ±m.
cm_log_reg = confusion_matrix(y_test, y_pred_log_reg)
```

```python
display_log_reg = ConfusionMatrixDisplay(confusion_matrix=cm_log_reg, display_labels=log_reg.classes_)
display_log_reg.plot()
plt.title("Confusion Matrix(KarmaÅŸÄ±klÄ±k Matrisi)")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy_log_reg = accuracy_score(y_test, log_reg.predict(X_test))
print("Lojistik Regresyon DoÄŸruluk:", accuracy_log_reg)
```
