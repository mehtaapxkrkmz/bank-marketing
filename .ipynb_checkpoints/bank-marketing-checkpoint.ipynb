```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
marketing = pd.read_csv('C:/Users/Asus/.cache/kagglehub/datasets/janiobachmann/bank-marketing-dataset/versions/1/bank.csv')
marketing
```

```python
#deposit,ürünü satın alıp almadığını ifade ediyor.Bizim bulmak istediğimiz
marketing['deposit']=marketing['deposit'].map({'yes': 1, 'no': 0})

# Bağımsız ve bağımlı değişkenleri ayırma
X = marketing.drop('deposit', axis=1)  # Tüm sütunlardan 'deposit' sütununu çıkarır.
y = marketing['deposit']              # 'deposit' sütununu hedef olarak seçilir.ç

X = X.drop('duration', axis=1)   #duration'u da çıkarıyorum.özel inceleme gerektiriyor.

# Kategorik sütunları one-hot encoding ile dönüştürme
X = pd.get_dummies(X, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
# Lojistik regresyon modelini oluşturup ve eğitme
log_reg = LogisticRegression(solver='liblinear',max_iter=1000,random_state=42)
```

```python
log_reg.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
log_reg.predict(X_test)
```

```python
# Test seti üzerinde tahminler.
y_pred_log_reg = log_reg.predict(X_test)

# Model performansını değerlendirme.Böylece modelin ne kadar doğru tahmin yaptığını anlayabiliriz.
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
report_log_reg = classification_report(y_test, y_pred_log_reg)

print("Accuracy(Doğruluk):", accuracy_log_reg)
print("Classification Report:\n", report_log_reg)

#support:sınıfın gerçek örnek sayısı.
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Tahmin edilen sınıflar ve gerçekteki sınıflar arasındaki karşılaştırmayı yapan "karışıklık matrisi"ni oluşturalım.
cm_log_reg = confusion_matrix(y_test, y_pred_log_reg)
```

```python
display_log_reg = ConfusionMatrixDisplay(confusion_matrix=cm_log_reg, display_labels=log_reg.classes_)
display_log_reg.plot()
plt.title("Confusion Matrix(Karmaşıklık Matrisi)")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy_log_reg = accuracy_score(y_test, log_reg.predict(X_test))
print("Lojistik Regresyon Doğruluk:", accuracy_log_reg)
```
